```python
import pandas as pd
import os
import time
from datetime import datetime
import numpy as np
```

```python
DATA_PATH1 = '/content/population.csv'
DATA_PATH2 = '/content/gdp.xlsx'
DATA_PATH3 = '/content/internet_users.json'
DATA_PATH4 = '/content/literacy_rate.xml'
OUT_DTR = 'Outputs'
INTERVAL_SECONDS = 20
MAX_RUNS = 1
```

```python
def extract():
    population_data = pd.read_csv(DATA_PATH1)
    population_data_len = len(population_data)
    print(f'Extracted {population_data_len} rows')
    #return population_data
    gdp_data = pd.read_excel(DATA_PATH2)
    gdp_data_len = len(gdp_data)
    print("extracted", gdp_data_len,"rows")
    #return gdp_data
    internet_users = pd.read_json(DATA_PATH3)
    internet_users_len = len(internet_users)
    print("extracted", internet_users, "rows")
    literacy_rate = pd.read_xml(DATA_PATH4)
    literacy_rate_len = len(literacy_rate)
    print("extracted", literacy_rate, "rows")
    return population_data,gdp_data,internet_users,literacy_rate



def transform(population_data,gdp_data,internet_users,literacy_rate):
    print("the first five rows  of population data is:", population_data.head())
    print("the shape of population data is:", population_data.shape)
    print("the type of population data is:", population_data.info())
    print("the column names of gdp is:", gdp_data.info())
    print("dataframe of internet users", internet_users)
    print("the dataframe of all required fields", literacy_rate)



```

```python
population_data,gdp_data,internet_users,literacy_rate = extract()
cleaned_data = transform(population_data,gdp_data,internet_users,literacy_rate)
```

```python
def transform1(population_data, gdp_data, internet_users, literacy_rate):
    datasets = {
        "Population": population_data,
        "GDP": gdp_data,
        "Internet Users": internet_users,
        "Literacy Rate": literacy_rate
    }

    for name, df in datasets.items():
        print(f"=== Exploring {name} Dataset ===")
        print(f"● Rows & Columns: {df.shape[0]} rows, {df.shape[1]} columns")
        print(f"● Column Names:  {df.columns.tolist()}")
        print(f"● Missing Values:\n{df.isnull().sum().to_string()}")
        print(f"● Duplicate Rows: {df.duplicated().sum()}")
        print("-" * 40)

    return population_data, gdp_data, internet_users, literacy_rate
_ = transform1(population_data, gdp_data, internet_users, literacy_rate)
```

```python
gdp_data = gdp_data.rename(columns={'Country': 'country'})


literacy_rate = literacy_rate.rename(columns={'name': 'country'})

print("Updated Column Names:")
print("GDP:", gdp_data.columns.tolist())
print("Literacy Rate:", literacy_rate.columns.tolist())
```

```python

population_data['population'] = pd.to_numeric(population_data['population'], errors='coerce')
gdp_data['GDP'] = pd.to_numeric(gdp_data['GDP'], errors='coerce')
internet_users['internet_users'] = pd.to_numeric(internet_users['internet_users'], errors='coerce')
literacy_rate['literacy_rate'] = pd.to_numeric(literacy_rate['literacy_rate'], errors='coerce')

print("Data types successfully validated and converted!")
```

```python

merged_df = pd.merge(population_data, gdp_data, on='country', how='inner')

merged_df = pd.merge(merged_df, internet_users, on='country', how='inner')

final_df = pd.merge(merged_df, literacy_rate, on='country', how='inner')


print(f"Successfully merged! Final DataFrame Shape: {final_df.shape}")
display(final_df.head())
```

```python
final_df['Internet Penetration (%)'] = round((final_df['internet_users'] / final_df['population']) * 100, 2)

final_df['GDP Per Capita'] = round(final_df['GDP'] / final_df['population'], 2)

print("Calculated columns added successfully!")
display(final_df[['country', 'population', 'internet_users', 'Internet Penetration (%)', 'GDP', 'GDP Per Capita']].head())
```

```python
import sqlite3

conn = sqlite3.connect('etl_database.db')

final_df.to_sql('Country_statistics', conn, if_exists='replace', index=False)

print("Database created and 'Country_statistics' table successfully populated!")

query = "SELECT * FROM Country_statistics LIMIT 5;"
verified_df = pd.read_sql(query, conn)

print("\nVerification - Reading back first 5 rows directly from SQLite:")
display(verified_df)
```

```python

conn = sqlite3.connect('etl_database.db')

print("--- 11. Top 5 Countries by GDP ---")
query_11 = "SELECT country, GDP FROM Country_statistics ORDER BY GDP DESC LIMIT 5;"
print(pd.read_sql(query_11, conn))
print("\n" + "="*50 + "\n")

print("--- 12. Average Literacy Rate ---")
query_12 = "SELECT AVG(literacy_rate) AS Average_Literacy_Rate FROM Country_statistics;"
print(pd.read_sql(query_12, conn))
print("\n" + "="*50 + "\n")

print("--- 13. Countries with Literacy Rate Above 90% ---")
query_13 = "SELECT country, literacy_rate FROM Country_statistics WHERE literacy_rate > 90;"
print(pd.read_sql(query_13, conn))
print("\n" + "="*50 + "\n")

print("--- 14. Top 5 Countries with Highest Internet Penetration ---")
query_14 = "SELECT country, `Internet Penetration (%)` FROM Country_statistics ORDER BY `Internet Penetration (%)` DESC LIMIT 5;"
print(pd.read_sql(query_14, conn))
print("\n" + "="*50 + "\n")

print("--- 15. Total Population ---")
query_15 = "SELECT SUM(population) AS Total_Population FROM Country_statistics;"
print(pd.read_sql(query_15, conn))
print("\n" + "="*50 + "\n")

print("--- 16. Country with the Highest GDP Per Capita ---")
query_16 = "SELECT country, `GDP Per Capita` FROM Country_statistics ORDER BY `GDP Per Capita` DESC LIMIT 1;"
print(pd.read_sql(query_16, conn))
print("\n" + "="*50 + "\n")

print("--- 17. Exporting Dataset to CSV ---")
final_df.to_csv('final_merged_countries.csv', index=False)
print("File successfully saved as 'final_merged_countries.csv'!")
print("\n" + "="*50 + "\n")

print("--- 18. Top 10 Countries by GDP Per Capita (Descending) ---")
query_18 = "SELECT country, `GDP Per Capita`, GDP, population FROM Country_statistics ORDER BY `GDP Per Capita` DESC LIMIT 10;"
print(pd.read_sql(query_18, conn))

conn.close()
```
